```python
#Converting created column to datetime
df['created'] = pd.to_datetime(df['created'])
df['date_created'] = df['created'].dt.date
```

```python
df.date_created
```

```python
df['year'] = pd.DatetimeIndex(df.date_created).year
df['month'] = pd.DatetimeIndex(df.date_created).month
df['day'] = pd.DatetimeIndex(df.date_created).day
```

```python
plt.figure(figsize=(10,5))
plt.xticks(fontsize = 10, rotation = 'vertical')
sns.countplot(x=df['year'], order = df.year.value_counts().sort_values(ascending = False).index)
plt.title('Movies per Year')
```

A greater part of the courses in this dataset was created in 2018

```python
top_10 = df.sort_values('num_subscribers', ascending=False).head(10)
top_10
```

```python
plt.figure(figsize=(6,15))
sns.barplot(y = 'title', x ='num_subscribers', data=top_10)
plt.title('Top 10 Courses with most subscriber')
plt.show()
```

From the above observation we can see 'An Entire MBA in 1 Course' is the most subscribed course

```python
plt.figure(figsize=(6,15))
plt.xticks(fontsize=10, rotation='vertical')
sns.barplot(x="num_published_lectures",y ="title",data = top_10)
plt.title('Top 10 Courses with most subscribed')
plt.show()
```

From the above observation we can see the course with most number of susbcribers is 'The Complete Finance Analyst Course 2020'.

```python
plt.figure(figsize=(6,15))
sns.barplot( x='price_detail__amount', y = 'title', data=top_10)
plt.title('Top 10 Courses by Price')
plt.show()
```

From the above observation we can see among the top 10 courses the priciest course is 'The Complete presentation and public speaking/speech course'.

```python
plt.figure(figsize=(6,5))
sns.barplot(x=top_10['title'], y='rating', data = top_10)
plt.xticks(fontsize = 10, rotation = 'vertical')
plt.title('Top 10 Courses by Rating')
plt.show()
```

From the above observation we can see the topmost rated course is 'The complete sql bootcamp 2020 from zero to hero'
